```python
#удаляем вин так как он нам не пригодится
data=data.drop(['vin'] , axis=1)

```

```python
#удаляем авто в которых нет информации о модели и марки
data=data.dropna(subset=['make'])
data=data.dropna(subset=['model'])
data.info()
```

```python
#удаляем авто в которых нет информации о трансмисии
data=data.dropna(subset=['transmission'])
data.info()
```

```python
#удаляем авто в которых нет информации о типе кузова
data=data.dropna(subset=['body'])
data.info()
```

```python
#удаляем авто в которых нет информации о пробеге
data=data.dropna(subset=['odometer'])
data.info()
```
